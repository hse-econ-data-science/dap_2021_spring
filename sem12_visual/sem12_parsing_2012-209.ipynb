#### ищем книжки на всех страницах

```python
from tqdm.notebook import tqdm
import time
time.sleep(3) # и пусть весь мир подождёт 3 секунды
```

```python
%%time
all_books = {}
for page in tqdm(range(1,50+1)):
    url_page = f'http://books.toscrape.com/catalogue/page-{page}.html'
    response = requests.get(url_page)
    tree = BeautifulSoup(response.content, 'html.parser')
    all_books_on_page = tree.find_all('h3')
    for book in all_books_on_page:
        link = book.a.get('href')
        title = book.a.get('title')
        all_books[link] = {'title':title}
```

```python
pd.DataFrame(all_books).T.head(10)
```

#### собираем инфу о книжках

```python
%%time
base_path = 'http://books.toscrape.com/catalogue/'
for book in tqdm(all_books):
    url_book = base_path + book
    response = requests.get(url_book)
    tree = BeautifulSoup(response.content, 'html.parser')
    tree.find_all('table')[0]
    b = pd.read_html(str(tree.find_all('table')[0]))[0]
    b = b.set_index(0)
    info_book = b.to_dict()[1]
    info_book['stars'] = tree.find_all('p', {'class':'star-rating'})[0].get('class')[1]
    all_books[book].update(info_book)
```

```python
df = pd.DataFrame(all_books).T
df.index.name = 'href'
df = df.reset_index()

df.head(10)
```

### joblib, try-except, tqdm

```python
from joblib import Parallel, delayed
```

```python
def get_more_info(book, base_path = 'http://books.toscrape.com/catalogue/'):
    url_book = base_path + book
    response = requests.get(url_book)
    tree = BeautifulSoup(response.content, 'html.parser')
    
    try:
        b = pd.read_html(str(tree.find_all('table')[0]))[0]
        b = b.set_index(0)
        info_book = b.to_dict()[1]
        info_book['stars'] = tree.find_all('p', {'class':'star-rating'})[0].get('class')[1]
        info_book['href'] = book
    except:
        print(url_book)
        info_book = dict()
        info_book['href'] = book
    return info_book

def update_info(book):
    all_books[book].update(get_more_info(book))
    return all_books[book]
```

```python
get_more_info(book)
```

```python
update_info(book)
```

```python
%%time

n_jobs = 16
r = Parallel(n_jobs=n_jobs)(
    delayed(update_info)(book) for book in tqdm(all_books)
)
```

```python
pd.DataFrame(r)
```

```python
tree = BeautifulSoup(response.content, 'html.parser')
table = tree.find_all('table', {'class':'items'})[0]
players = table.find_all('tr', {'class':["odd",'even']})
len(players)
```

```python
def get_player_info(player):
    result = {
        'name': player.find_all('span', {'class':'hide-for-small'})[-1].get_text(),
        'price_old': player.find_all('td', {'class':'rechts hauptlink'})[0].span.get('title'),
        'price':player.find_all('td', {'class':'rechts hauptlink'})[0].get_text().replace(u'\xa0', u'')
    }
    return result
```

```python
res = []
for player in players:
    player_price = get_player_info(player)
    res.append(player_price)

pd.DataFrame(res)
```

```python
search_term = 'корол шут кукла'
```

```python
querystring = f"http://api.genius.com/search?q={urllib.quote(search_term)}"
request = urllib.Request(querystring)
request.add_header("Authorization", "Bearer " + client_access_token)
request.add_header(
    "User-Agent",
    "curl/7.9.8 (i686-pc-linux-gnu) libcurl 7.9.8 (OpenSSL 0.9.6b) (ipv6 enabled)"
)  #Must include user agent of some sort, otherwise 403 returned
```
