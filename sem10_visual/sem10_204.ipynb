```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['axes.titlesize'] = 18
rcParams['figure.figsize'] = (8,2)

import seaborn as sns

import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, num=1000)
y = np.sin(x)
z = np.cos(x)

plt.figure(figsize=(8,2))
plt.plot(x, y, color=(0, 0.5, 0.3), linestyle='--', label='синус')
plt.plot(x, z, color='magenta', lw=3, label='косинус')
plt.fill_between(x, y, z,
        color='yellow', alpha=0.2)
    
plt.title('Синус и косинус')
plt.xlabel('Ось иксов')
plt.ylabel('Ось игреков')
plt.legend(fontsize=12, loc='lower left');
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df[['total_bill', 'tip']].hist(
    figsize=(8,2), color='red',
    density=True, bins=100);
```

```python
df['sex'].value_counts().plot(kind='barh');
```

```python
sns.jointplot(data = df, 
              x = 'total_bill',
              y = 'tip', kind = 'hex');
```

```python
sns.scatterplot(data = df, 
              x = 'total_bill',
              y = 'tip');
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x = 'day', y = 'total_bill', 
            hue = 'sex', data = df);
```
