```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

rcParams['figure.figsize'] = (8, 2)
rcParams['axes.titlesize'] = 20

# Какие вообще опции быват
# rcParams.keys()

import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 300)
y1 = np.cos(x)
y2 = np.sin(x)

#plt.figure(figsize = (8,2))
plt.plot(x, y1, color='blue', label='cos(x)')
plt.plot(x, y2, color=(0.1, 0.5, 0), 
         linestyle='dotted', linewidth=3, label='sin(x)')

plt.fill_between(x, y2, y1, color='magenta', 
                 alpha=0.1, label='межсинусье')

plt.title('Графический график')
plt.xlabel('ось иксов', fontsize=14)
plt.ylabel('ось игреков', fontsize=14)
plt.legend(fontsize=10, loc='lower right');
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
ax = df[['total_bill', 'tip']].hist()

ax[0,0].set_title('ацоцауощ')
ax[0,1].set_title('лоулук');
```

```python
fig, ax = plt.subplots(2, 2, figsize=(8,8))

ax[1,1].plot(x, y1)
ax[1,1].set_title('синус')

df['day'].value_counts().plot(kind='bar', ax=ax[0,0])

sns.kdeplot(df['total_bill'], ax=ax[0,1])
sns.rugplot(df['total_bill'], ax=ax[0,1]);

sns.scatterplot(data=df, x='total_bill', y='tip', ax=ax[1,0]);
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x = 'day', y = 'total_bill', data = df);
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x = 'day', y = 'total_bill', hue='sex', data = df);
```

```python
df.day.value_counts()
```

```python
sns.stripplot(data = df, x = 'day',
              y = 'total_bill', jitter = True, hue = 'sex');
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(df.corr(), cmap='rainbow',
            vmin=-1, vmax=1, 
            square=True,
            linewidths=0.1,annot_kws={"fontsize":22},
            annot=True);
```
